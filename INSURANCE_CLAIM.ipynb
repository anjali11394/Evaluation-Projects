#TREATING OUTLIERS

```python
#removing of outliers

#removing outliers

q1=df.quantile(0.25)

q3=df.quantile(0.75)

iqr=q3=q1
```

```python
#validating outlier
property_claim_high=(q3.property_claim + (1.5 * iqr.property_claim))
print(property_claim_high)


#checking the indexes with higher values
index=np.where(df['property_claim']>property_claim_high)
print(index)

#dropping the index which have high values
df=df.drop(df.index[index])
df.shape

#reseting the data after dropping outliers
df.reset_index()
```

```python
#validating outlier
policy_annual_premium_high=(q3.policy_annual_premium + (1.5 * iqr.policy_annual_premium))
print(policy_annual_premium_high)


#checking the indexes with higher values
index=np.where(df['policy_annual_premium']>policy_annual_premium_high)
print(index)

#dropping the index which have high values
df=df.drop(df.index[index])
df.shape

#reseting the data after dropping outliers
df.reset_index()
```

```python
#validating outlier
policy_annual_premium_high=(q1.policy_annual_premium - (1.5 * iqr.policy_annual_premium))
print(policy_annual_premium_high)


#checking the indexes with higher values
index=np.where(df['policy_annual_premium'] <policy_annual_premium_high)
print(index)

#dropping the index which have high values
df=df.drop(df.index[index])
df.shape

#reseting the data after dropping outliers
df.reset_index()
```

#MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**LOGISTIC REGRESSION**

```python
lr=LogisticRegression()
```

```python
for i in range(0,1000):
  x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=i)
  lr.fit(x_train,y_train)
  pred_train=lr.predict(x_train)
  pred_test=lr.predict(x_test)
  if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
    print('At random state',i,'the model performs well')
    print('At random state',i)
    print('Training accuracy of data is :-',accuracy_score(y_train,pred_train)*100)
    print('Testing accuracy of data is :-',accuracy_score(y_test,pred_test)*100)
    print('\n')
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=27)
```

**CLASSIFICATION REPORT(LOGISTIC REGRESSION)**

```python
print(classification_report(y_test,pred_test))
```

**cross-val(logistic regression)**

```python
from sklearn.model_selection import cross_val_score
```

```python
pred_lr=lr.predict(x_test)

lss=accuracy_score(y_test,pred_lr)
for j in range(2,10):
  lsscore=cross_val_score(lr,X,y,cv=j)
  lsc=lsscore.mean()
  print('At cv:-',j)
  print('cross validation score is :-',lsc*100)
  print('accuracy score is:-',lss*100)
  print('\n')
```

```python
#cv score at 8 has least differnce b/w accuracy and cv score

lsscore_selected=cross_val_score(lr,X,y,cv=8).mean()
print('The cv score is ',lsscore_selected*100, '\n the accuracy score is ',lss*100)
```
